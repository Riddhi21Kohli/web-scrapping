```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_product_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    product_description = soup.find('div', {'id': 'productDescription'})
    if product_description:
        product_description = product_description.text.strip()
    else:
        product_description = ""


    asin = soup.find('th', text='ASIN')
    if asin:
        asin = asin.find_next('td').text.strip()
    else:
        asin = ""

  
    manufacturer = soup.find('a', {'id': 'bylineInfo'})
    if manufacturer:
        manufacturer = manufacturer.text.strip()
    else:
        manufacturer = ""

    return product_description, asin, manufacturer


def scrape_product_listing_pages():
    base_url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_"
    total_pages = 20
    product_count = 0
    page_count = 1

    csv_headers = ['Product URL', 'Product Name', 'Product Price', 'Rating', 'Number of Reviews',
                   'Description', 'ASIN', 'Product Description', 'Manufacturer']

    with open('product_data.csv', 'w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(csv_headers)

        while page_count <= total_pages and product_count < 200:
            url = base_url + str(page_count)
            response = requests.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            products = soup.find_all('div', {'data-component-type': 's-search-result'})

            for product in products:
                if product_count >= 200:
                    break

                # Extracting product details
                product_url = "https://www.amazon.in" + product.find('a', class_='a-link-normal')['href']
                product_name = product.find('span', class_='a-size-medium').text.strip()
                product_price = product.find('span', class_='a-price-whole').text.strip()
                rating = product.find('span', class_='a-icon-alt').text.split()[0]
                reviews = product.find('span', {'class': 'a-size-base'}).text.strip()

                
                product_description, asin, manufacturer = scrape_product_details(product_url)

               
                writer.writerow([product_url, product_name, product_price, rating, reviews,
                                 product_description, asin, product_description, manufacturer])

                product_count += 1

            page_count += 1

    print("Data scraping and export complete.")


scrape_product_listing_pages()
```
